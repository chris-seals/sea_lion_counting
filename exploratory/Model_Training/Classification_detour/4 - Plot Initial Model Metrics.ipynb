```python
## TODO : Add a few lines to end of each modeling notebook to record scores/model//parameters? That way fresh visualization is \
## instantaneous.
f1 = pd.read_excel('../../results/model_results_balanced_classes.xlsx',sheet_name=0)
accuracy = pd.read_excel('../../results/model_results_balanced_classes.xlsx',sheet_name=1)
```
