#### Floodfill Test

```python
import cv2
import numpy as np

# read image
filename = path1
im = cv2.imread(filename)
h,w = im.shape[:2]

# flood fill example
diff = (6,6,6)
mask = np.zeros((h+2,w+2),np.uint8)
cv2.floodFill(im, mask, (60,60), (145,255,0), diff, diff)#@, cv2.FLOODFILL_FIXED_RANGE)


# save the result


plt.imshow(im)
```
